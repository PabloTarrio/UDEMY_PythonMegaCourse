```python
# Si queremos que el data frame tenga ese cambio de INDEX permanente, tenemos dos opciones:
#     1. Crear un nuevo dataframe con el index en una nueva variable
```

```python
df9 = df8.set_index("ID")
df9
```

```python
#     2. Utilizar el par√°metro "inplace=True" al usar el m√©todo set_index()
# Esta segunda opci√≥n tiene el inconveniente de que al establecer una columna como index no devuelve a columna el index que est√° puesto,
# simplemente lo elimina. Por ejemplo, si usasemos "Address" como index, la columna "ID" desaparecer√≠a.
```

```python
df8.set_index("ID", inplace=True)
df8
```
