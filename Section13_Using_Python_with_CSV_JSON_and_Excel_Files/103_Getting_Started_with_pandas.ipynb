Importamos la librería PANDAS: Librería que proporciona estructuras de datos y herramientas para análisis de datos en Python

```python
import pandas as pd
```

Creamos el objeto df1 (DATAFRAME) que contiene los datos

```python
df1 = pd.DataFrame([[2,4,6],[10,20,30]])
```

Imprimimos los datos que hemos guardado en el dataframe df1

```python
df1
```

Cambiar el nombre de las columnas  

```python
df1 = pd.DataFrame ([[2,4,6],[10,20,30]], columns=["Prize","Age","Value"] )
df1
```

```python
df1 = pd.DataFrame ([[2,4,6],[10,20,30]], columns=["Prize","Age","Value"], index=["First","Second"] )
df1
```

Creamos un nuevo DataFrame, pero incluimos diccionario como datos.

```python
df2 = pd.DataFrame([{"Name":"John"},{"Name":"Jack"}])
df2
```
